##pre-filtering
- now we will remove all the users that have related less movies and

- also all those movies that has been rated less in no.s

```python
dataset_movie_summary= netflix_dataset.groupby('Movie_id')['Rating'].agg(['count'])

dataset_movie_summary
```

```python
# now we will create a benchmark

movie_brenchmark= round(dataset_movie_summary['count'].quantile(0.6),0)
movie_brenchmark
```

```python
drop_movie_list= dataset_movie_summary[dataset_movie_summary['count']<movie_brenchmark].index
drop_movie_list
```

```python
len(drop_movie_list)
```

```python
4499-2699
```

```python
#now we will remove all the users that are in-active
dataset_cust_summary= netflix_dataset.groupby('Cust_id')['Rating'].agg(['count'])
dataset_cust_summary
```

```python
cust_brenchmark= round(dataset_cust_summary['count'].quantile(0.6),0)
cust_brenchmark
```

```python
drop_cust_list= dataset_cust_summary[dataset_cust_summary['count']< cust_brenchmark].index
drop_cust_list
```

```python
len(drop_cust_list)
```

##Lets remove all movies that has been rated less than 908 times

##Lets remove all customers that have rated less than 36 movies

```python
netflix_dataset= netflix_dataset[~netflix_dataset['Movie_id'].isin(drop_movie_list)]
netflix_dataset= netflix_dataset[~netflix_dataset['Cust_id'].isin(drop_cust_list)]
```

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
reader=Reader()
```

```python
data= Dataset.load_from_df(netflix_dataset[['Cust_id','Movie_id','Rating']][:200000],reader)
```

```python
model=SVD()
```

```python
cross_validate(model,data, measures=['RMSE'], cv=3, verbose= True)
```
